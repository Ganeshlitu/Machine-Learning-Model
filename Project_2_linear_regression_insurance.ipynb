```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('/content/insurance.csv')
```

```python
data.head()
```

```python
data['sex']=data['sex'].apply({'female':0,'male':1}.get)
data['smoker']=data['smoker'].apply({'yes':1,'no':0}.get)
data['region']=data['region'].apply({'southwest':1,'northwest':2,'southeast':3,'northeast':4}.get)
```

```python
data.head()
```

```python
x=data.drop('expenses',axis=1)
y=data['expenses']
```

```python
from sklearn.model_selection import train_test_split
x_test,x_train,y_test,y_train= train_test_split(x,y,train_size=0.8)
```

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
prediction= regression.predict(x_test)
```

```python
#to check accuracy
regression.score(x,y)
```
