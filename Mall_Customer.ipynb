```python
import numpy as np
import pandas as pd
```

```python
data_new=pd.read_csv("/content/Mall_Customers.csv")
```

```python
data_new.head()
```

```python
data=data_new.drop(['CustomerID'], axis=1)
```

```python
data['Gender']=data['Gender']. apply({'Male':0,'Female':1}.get)
```

```python
data.head()
```

```python
from sklearn.cluster import DBSCAN
```

```python
cluster=DBSCAN(eps=5,min_samples=5).fit(data)
```

```python
DBSCAN_dataset=data.copy()
DBSCAN_dataset.loc[:,'cluster']=cluster.labels_
```

```python
DBSCAN_dataset.cluster.value_counts().to_frame()
```
