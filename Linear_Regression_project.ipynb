Gathering the data

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("/content/Housing.csv")
```

```python
data.head()
```

```python
#Data processing - converting categorical variable to numeric using one hot encoding technique
```

```python
data["mainroad"]=data["mainroad"].apply({"yes":1,"no":0}.get)
data["basement"]=data["basement"].apply({"yes":1,"no":0}.get)
data["hotwaterheating"]=data["hotwaterheating"].apply({"yes":1,"no":0}.get)
data["airconditioning"]=data["airconditioning"].apply({"yes":1,"no":0}.get)
data["prefarea"]=data["prefarea"].apply({"yes":1,"no":0}.get)
data["furnishingstatus"]=data["furnishingstatus"].apply({"furnished":1,"semi-furnished":2,"unfurnished":0}.get)
data["guestroom"]=data["guestroom"].apply({"yes":1,"no":0}.get)
```

```python
data.head()
```

```python
x=data.drop("price",axis=1)
y=data["price"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size= 0.6)
```

Creating MLM using Liner regression algorithm

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression() #activating Linear regression
```

```python
regression.fit(x_train,y_train) #.fit is used for training
```

```python
predictions=regression.predict(x_test) #.predict is used to predict the future value
```

```python
#accuracy
regression.score(x,y)
```
